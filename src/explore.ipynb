```python
import pandas as pd

DATABASE_URL="https://raw.githubusercontent.com/4GeeksAcademy/k-means-project-tutorial/main/housing.csv"

total_data = pd.read_csv(DATABASE_URL)

total_data.head()
```

<b>Step 1: Loading the dataset</b>

The dataset can be found in this project folder under the name housing.csv. You can load it into the code directly from the link (https://raw.githubusercontent.com/4GeeksAcademy/k-means-project-tutorial/main/housing.csv) or download it and add it by hand in your repository. In this case we are only interested in the Latitude, Longitude and MedInc columns.

```python
columns_to_drop = ['HouseAge', 'AveRooms', 'AveBedrms', 'Population', 'AveOccup', 'MedHouseVal']

total_data.drop(columns=columns_to_drop, inplace=True)
```

```python
total_data.head()
```

```python
total_data.shape
```

Be sure to conveniently split the dataset into train and test as we have seen in previous lessons. Although these sets are not used to obtain statistics, you can use them to train the unsupervised algorithm and then to make predictions about new points to predict the cluster they are associated with.

```python
from sklearn.model_selection import train_test_split
X = total_data

X_train, X_test = train_test_split(X, test_size = 0.2, random_state = 42)

X_train.head()
```

<b>Step 2: Build a K-Means</b>

Classify the data into 6 clusters using the K-Means model. Then store the cluster to which each house belongs as a new column in the dataset. You could call it cluster. To introduce it to your dataset you may have to categorize it. See what format and values it has and act accordingly. Plot it in a dot plot and describe what you see.

```python
import numpy as np
from sklearn.cluster import KMeans
# from sklearn.datasets import make_blobs
```

```python
model_unsup = KMeans(n_clusters = 6, n_init = "auto", random_state = 42)
model_unsup.fit(X_train)
```

```python
y_train = list(model_unsup.labels_)
"""This line is extracting the cluster labels from an unsupervised learning model (model_unsup) and converting them into a list. 
In unsupervised learning, particularly clustering algorithms like K-means, the model assigns each data point to a cluster. 
These cluster labels are stored in the labels_ attribute of the fitted model object. 
By calling model_unsup.labels_, you're accessing these cluster labels."""

y_train
```

```python
X_train["cluster"] = y_train
""" add a new column to X_train. This column will contain the cluster labels from y_train corresponding to each data point. """
X_train.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

fig, axis = plt.subplots(1, 3, figsize = (15, 5))

sns.scatterplot(ax = axis[0], data = X_train, x = "Latitude", y = "Longitude", hue = "cluster", palette = "pastel")
sns.scatterplot(ax = axis[1], data = X_train, x = "Latitude", y = "MedInc", hue = "cluster", palette = "pastel")
sns.scatterplot(ax = axis[2], data = X_train, x = "Longitude", y = "MedInc", hue = "cluster", palette = "pastel")
plt.tight_layout()

plt.show()
```

<b>Step 3: Predict with the test set</b>

Now use the trained model with the test set and add the points to the above plot to confirm that the prediction is successful or not.

```python
y_test = list(model_unsup.predict(X_test))
```

```python
y_test
```

```python
X_test["cluster"] = y_test
X_test
```

```python
fig, axis = plt.subplots(1, 3, figsize = (15, 5))

sns.scatterplot(ax = axis[0], data = X_train, x = "Latitude", y = "Longitude", hue = "cluster", palette = "pastel")
sns.scatterplot(ax = axis[1], data = X_train, x = "Latitude", y = "MedInc", hue = "cluster", palette = "pastel")
sns.scatterplot(ax = axis[2], data = X_train, x = "Longitude", y = "MedInc", hue = "cluster", palette = "pastel")

sns.scatterplot(ax = axis[0], data = X_test, x = "Latitude", y = "Longitude", hue = "cluster", palette = "pastel")
sns.scatterplot(ax = axis[1], data = X_test, x = "Latitude", y = "MedInc", hue = "cluster", palette = "pastel")
sns.scatterplot(ax = axis[2], data = X_test, x = "Longitude", y = "MedInc", hue = "cluster", palette = "pastel")

plt.tight_layout()

plt.show()
```

<b>Step 4: Train a supervised classification model</b>

Now that K-Means has returned a categorization (clustering) of the points for the training and test sets, study which model might be most useful and train it. Get the statistics and describe what you see.

This flow is very common when we have unlabeled data: use an unsupervised learning model to label it automatically and then a supervised learning model.

```python
from sklearn.tree import DecisionTreeClassifier

model_sup = DecisionTreeClassifier(random_state = 42)
model_sup.fit(X_train, y_train)
```

```python
from sklearn import tree

fig = plt.figure(figsize=(25,25))

tree.plot_tree(model_sup, feature_names = list(X_train.columns), class_names = ["0", "1", "2", "3", "4", "5"], filled = True)

plt.show()
```

```python
y_pred = model_sup.predict(X_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score

accuracy_score(y_test, y_pred)
```
